```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

# Set a clean visual theme for the plots
sns.set_theme(style="whitegrid")

# Load the Iris dataset directly via seaborn
iris_df = sns.load_dataset('iris')

print(f"Dataset Shape: {iris_df.shape}\n")
print(f"Columns: {list(iris_df.columns)}\n")
print("First 5 rows:")
iris_df.head()

print("--- Dataset Information ---")
iris_df.info()

print("\n--- Descriptive Statistics ---")
iris_df.describe()

plt.figure(figsize=(8, 5))
sns.scatterplot(data=iris_df, x='sepal_length', y='sepal_width', hue='species', palette='Set1')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

sns.histplot(data=iris_df, x='sepal_length', hue='species', kde=True, ax=axes[0, 0], palette='Set2')
axes[0, 0].set_title('Sepal Length Distribution')

sns.histplot(data=iris_df, x='sepal_width', hue='species', kde=True, ax=axes[0, 1], palette='Set2')
axes[0, 1].set_title('Sepal Width Distribution')

sns.histplot(data=iris_df, x='petal_length', hue='species', kde=True, ax=axes[1, 0], palette='Set2')
axes[1, 0].set_title('Petal Length Distribution')

sns.histplot(data=iris_df, x='petal_width', hue='species', kde=True, ax=axes[1, 1], palette='Set2')
axes[1, 1].set_title('Petal Width Distribution')

plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
# Melting shifts the table format so we can view all features side-by-side
melted_df = pd.melt(iris_df, id_vars='species', value_vars=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])

sns.boxplot(data=melted_df, x='variable', y='value', hue='species', palette='Pastel1')
plt.title('Feature Distributions & Outliers by Species')
plt.xlabel('Features')
plt.ylabel('Values (cm)')
plt.show()
```
